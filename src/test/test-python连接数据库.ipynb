```python
import pymysql
from pymysql.cursors import DictCursor

# 建立数据库连接
connection = pymysql.connect(
    host='localhost',      # 数据库主机地址
    user='root',  # 数据库用户名
    password='123456',  # 数据库密码
    database='ai_edu',  # 数据库名
    charset='utf8mb4'      # 字符集
)
try:
    # 创建游标对象
    with connection.cursor() as cursor:
        # 执行SQL查询
        sql = """
                select
                    user_id
                from
                    test_exam_question
        """
        cursor.execute(sql)

        # 使用 fetchall() 获取所有查询结果
        results = cursor.fetchall()
        print(results)

        # # 遍历并处理结果
        # for row in results:
        #     user_id = row[0]
        #     name = row[1]
        #     age = row[2]
        #     print(f"ID: {user_id}, Name: {name}, Age: {age}")
finally:
    # 关闭数据库连接
    connection.close()
```
